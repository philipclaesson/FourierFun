```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plotWave(periods, wave, title=""): 
    plt.plot(wave)
    plt.title(title)
    plt.ylabel('Amplitude')
    plt.xlabel('Time')
    plt.axis([0, periods*period/sample_rate, -1, 1])
    plt.show()
```

```python
## Sine Wave

period = 2*np.pi # the period is 2pi
sample_rate = 0.1 # sample rate
time = np.arange(0, 2 * period, sample_rate);
sineWave = np.sin(time)

plotWave(2, sineWave, "Sine Wave")
```

```python
# Cosine Wave
cosWave = np.cos(time)
plotWave(2, cosWave, "Cosine Wave")
```

### Fourier Series
A sawtooth can be approximated through the fourier series
![Sawtooth](img/sawtooth.png)

So let's do that. 

```python
# Sawtooth wave
A = 1
T = 2*np.pi # the period is 2pi
sample_rate = 0.1 # sample rate
time = np.arange(0, 4 * T, sample_rate);
plot_these = [3, 10, 100]


sawtooth = np.zeros(time.shape)
for m in range(101)[1:]:
    for i, t in enumerate(time):
        sawtooth[i] -= (2 * A * np.cos(np.pi * m))/(np.pi*m) * np.sin(2 * np.pi * m * t / T)
    if (m in plot_these): 
        plt.plot(sawtooth, label = m)

            
plt.plot(np.zeros(time.shape), 'r--', linewidth = 0.9)
plt.title("Sawtooth")
plt.ylabel('Amplitude')
plt.xlabel('Time')
plt.legend()
plt.axis([0, 4*T/sample_rate, -1.5, 1.5])
plt.show()
```
